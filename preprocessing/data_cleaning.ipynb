```python
import pandas as pd

# Load the CSV file
result = pd.read_csv("data/personal.csv")

# Remove the '观影人数' column
result = result.drop(columns=['观影人数', 'Out'])

# Rename the specified columns
result = result.rename(columns={
    '种类': 'media_type',
    '状态': 'status',
    '完成日期': 'completed_date',
    '评级': 'rating',
    '短评': 'comment',
    'Brief Desc': 'summary'
})

# Filter rows where status is 'Completed'
completed_rows = result[result['status'] == 'Completed']
```
